## Deriving Data Solutions as per Customer Requirements
### The solutions are based on what questions an average customer can ask. Some of them are given below:
1) Let's say a client wants to install its carrier and they have given the specific latitude, longitude. They can ask for locations near those given coordinates <br>
2) They can ask for tower type they should go for to get maximum results

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:/INTERNSHIP/db_final.csv")
df.head(10)
```

### The statistics for all the types of towers are given below

```python
types = ["self-supporting lattice", "guyed lattice", "utility pole", "monopole", "power mount", "rooftop"]

for tower_type in types:
    print(tower_type)
    print(df[df["Tower Type"]==tower_type]["Carriers"].describe())
    print(df[df["Tower Type"]==tower_type]["Height"].describe())
```

### Find the closest coordinates within 5 kim radius from the centre(given coordinates) 

```python
#Find the closest coordinates to the given coordinates
def closest_coordinates(lat, long):
    
    coord_diff = (df["Latitude"]-lat).abs() + (df["Longitude"]-long).abs()    #absolute difference between the 2 coordinates
    coord_sorted = coord_diff.argsort()    # sort the coordinates based on closeness to give coordinates
    return df.iloc[coord_sorted[:5]]     #Choose the 10 closest coordinates
    
#Given coordinates
lat = 41.50
long = -72.3
radius_limit = 5000   # 5 km radius from the centre which is the given latitude and longitude
df_closest = closest_coordinates(lat, long)
df_closest
```

### Let's say for example, T-Mobile Internet Service Provider asks for locations near the given location. The data solution to this question would be the coordinates (41.487, -72.313) and (41.471, -72.323) because the other 3 towers already have T-Mobile as a carrier. <br>
### T-Mobile also wants to know which type of tower they should go for to get maximum results. The data solution to this question would be the tower type "Monopole" as it is the most popular tower type in that area.
